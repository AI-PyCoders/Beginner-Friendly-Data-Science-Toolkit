```python
import sys
import os
import pandas as pd

# Add the path to the scripts directory
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../scripts')))

from cleaning import handle_missing_values, remove_duplicates
from eda import summary_statistics, plot_correlation_matrix
from visualizations import plot_histogram, plot_scatter

# Load the datasets
df = pd.read_csv("../data/iris.csv")

# Inspect the DataFrame
print(df.head())
print(df.info())

# Convert all numeric columns to numeric types
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Data Cleaning
df = handle_missing_values(df, strategy="mean")
df = remove_duplicates(df)

# EDA
summary_statistics(df)
plot_correlation_matrix(df)
plot_histogram(df, column='SepalLengthCm')
plot_scatter(df, x='SepalLengthCm', y='SepalWidthCm')

```
